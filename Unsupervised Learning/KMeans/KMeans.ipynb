```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

- Data Hazırlama

```python
# class1
x1 = np.random.normal(25, 5, 1000)
y1 = np.random.normal(25, 5, 1000)

# class2
x2 = np.random.normal(55, 5, 1000)
y2 = np.random.normal(60, 5, 1000)

# class3
x3 = np.random.normal(55, 5, 1000)
y3 = np.random.normal(15, 5, 1000)


x = np.concatenate((x1, x2, x3), axis = 0)
y = np.concatenate((y1, y2, y3), axis = 0)

dictionary = {"x": x, "y": y}

data = pd.DataFrame(dictionary)
```

```python
data.head()
```

```python
plt.scatter(x1, y1)
plt.scatter(x2, y2)
plt.scatter(x3, y3)
```

- Verimizde böyle görünecek

```python
plt.scatter(x1, y1, color = "black")
plt.scatter(x2, y2, color = "black")
plt.scatter(x3, y3, color = "black")
```

- KMeans find n value

```python
from cv2 import kmeans
from sklearn.cluster import KMeans

wcss = []

for i in range(1, 15):
    kmean = KMeans(n_clusters = i)
    kmean.fit(data)
    wcss.append(kmean.inertia_)

plt.plot(range(1, 15), wcss)
plt.xlabel("number of k (cluster) value")
plt.ylabel("wcss")
plt.title("KMeans")
plt.show()
```

```python
kmeans2 = KMeans(n_clusters = 3)
clusters = kmeans2.fit_predict(data)

data["label"] = clusters
plt.scatter(data.x[data.label == 0], data.y[data.label == 0], color = "green")
plt.scatter(data.x[data.label == 1], data.y[data.label == 1], color = "blue")
plt.scatter(data.x[data.label == 2], data.y[data.label == 2], color = "red")
plt.show()
```
