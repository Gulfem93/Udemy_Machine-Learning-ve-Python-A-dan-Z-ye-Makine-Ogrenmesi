```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

- Data Hazırlama

```python
# class1
x1 = np.random.normal(25, 5, 100)
y1 = np.random.normal(25, 5, 100)

# class2
x2 = np.random.normal(55, 5, 100)
y2 = np.random.normal(60, 5, 100)

# class3
x3 = np.random.normal(55, 5, 100)
y3 = np.random.normal(15, 5, 100)


x = np.concatenate((x1, x2, x3), axis = 0)
y = np.concatenate((y1, y2, y3), axis = 0)

dictionary = {"x": x, "y": y}

data = pd.DataFrame(dictionary)
```

```python
data.head()
```

```python
plt.scatter(x1, y1)
plt.scatter(x2, y2)
plt.scatter(x3, y3)
```

- Verimizde böyle görünecek

```python
plt.scatter(x1, y1, color = "black")
plt.scatter(x2, y2, color = "black")
plt.scatter(x3, y3, color = "black")
```

- Dendogram

```python
from scipy.cluster.hierarchy import linkage, dendrogram

merg = linkage(data, method = "ward")
dendrogram(merg, leaf_rotation = 90)

plt.xlabel("Data Point")
plt.ylabel("Euclidean Distance")
plt.show()
```

- Hierarchical Clustering (HC)

```python
from sklearn.cluster import AgglomerativeClustering

hierarchical_clustering = AgglomerativeClustering(n_clusters = 3, affinity = "euclidean", linkage = "ward")
clustering = hierarchical_clustering.fit_predict(data)
```

```python
data["label"] = clustering
plt.scatter(data.x[data.label == 0], data.y[data.label == 0], color = "green")
plt.scatter(data.x[data.label == 1], data.y[data.label == 1], color = "blue")
plt.scatter(data.x[data.label == 2], data.y[data.label == 2], color = "red")
plt.show()
```
